```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 as db
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
from pandasql import sqldf

# Read the data set 

sqladb=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',names=['Age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race','sex','capitalgain','capitalloss','hoursperweek','nativecountry','>50K, <=50K'])
```

```python
pysqldf("SELECT * FROM sqladb LIMIT 10;")
```
